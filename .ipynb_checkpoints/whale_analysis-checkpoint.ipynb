# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:

* `whale_returns.csv`: Contains returns of some famous "whale" investors' portfolios.

* `algo_returns.csv`: Contains returns from the in-house trading algorithms from Harold's company.

* `sp_tsx_history.csv`: Contains historical closing prices of the S&P TSX 60 Index.

## Whale Returns

Read the Whale Portfolio daily returns and clean the data.

```python
# Reading whale returns
whale_returns_csvpath = Path("./Resources/whale_returns.csv")
whale_returns_df = pd.read_csv(whale_returns_csvpath, parse_dates=True, index_col="Date")
```

```python
# Show information about the whale_returns data frame
whale_returns_df.info()
```

```python
# Show the head and tail of the data frame
whale_returns_df
```

```python
# Count rows with nulls by summing the results of isnull (which returns True as 1) over the dataframe, even though this is already indicated in the .info call
print("Number of nulls per column")
whale_returns_df.isnull().sum()
```

```python
# Drop nulls
whale_returns_df = whale_returns_df.dropna().copy()
```

```python
# Show the head and tail of the data frame - rows with NaN should now be gone
whale_returns_df     # alternatively could have used whale_returns_df.head() and whale_returns_df.tail()
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data.

```python
# Reading algorithmic returns
algo_returns_csvpath = Path("./Resources/algo_returns.csv")
algo_returns_df = pd.read_csv(algo_returns_csvpath, parse_dates=True, index_col="Date") # Read the CSV into a data frame while converting the Date column into the index and dropping it
```

```python
# Convert the dates to a DateTimeIndex
# Completed as part of previous step
algo_returns_df
```

```python
# Count nulls
print("Number of nulls per column")
algo_returns_df.isnull().sum()
```

```python
# Drop nulls
algo_returns_df = algo_returns_df.dropna().copy()
```

## S&P TSX 60 Returns

Read the S&P TSX 60 historic closing prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P TSX 60 Closing Prices
sptsx60_returns_csvpath = Path("./Resources/sp_tsx_history.csv")
sptsx60_returns_df = pd.read_csv(sptsx60_returns_csvpath, parse_dates=True, index_col="Date") # Read the CSV into a data frame while converting the Date column into the index and dropping it
```

```python
# Convert the dates to a DateTimeIndex
# Completed as part of previous step
sptsx60_returns_df
```

```python
# Check Data Types
sptsx60_returns_df.dtypes
```

```python
# Fix Data Types
sptsx60_returns_df["Close"] = sptsx60_returns_df["Close"].str.replace("$", "")   # Remove any dollar signs ($) from the Close column
sptsx60_returns_df["Close"] = sptsx60_returns_df["Close"].str.replace(",", "")   # Remove any commas (,) from the Close column
sptsx60_returns_df["Close"] = sptsx60_returns_df["Close"].astype("float")        # Convert the Close column to a float data type
sptsx60_returns_df
```

```python
# Calculate Daily Returns
sptsx60_daily_returns = sptsx60_returns_df.pct_change()
sptsx60_daily_returns.head
```

```python
# Drop nulls
sptsx60_daily_returns = sptsx60_daily_returns.dropna().copy()
sptsx60_daily_returns.head
```

```python
# Rename `Close` Column to be specific to this portfolio.
sptsx60_daily_returns = sptsx60_daily_returns.rename(columns={"Close": "S&P TSX"})  # Rename the Close column to S&P TSX
```

## Combine Whale, Algorithmic, and S&P TSX 60 Returns

```python
# Join Whale Returns, Algorithmic Returns, and the S&P TSX 60 Returns into a single DataFrame with columns for each portfolio's returns.
joined_df = pd.concat([whale_returns_df, algo_returns_df, sptsx60_daily_returns], axis="columns", join="inner")
joined_df
```
